```python
from genopandas import AnnotatedMatrix

df = pd.DataFrame({
        'sample_1': [1, 2, 3],
        'sample_2': [4, 5, 6]
    },
    index=['gene_a', 'gene_b', 'gene_c'])

sample_data = pd.DataFrame(
    {'condition': ['control', 'treated']},
    index=['sample_1', 'sample_2'])

matrix = AnnotatedMatrix(df, sample_data=sample_data)
matrix
```
