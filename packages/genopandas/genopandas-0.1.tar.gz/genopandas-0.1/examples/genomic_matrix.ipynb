```python
import numpy as np
import pandas as pd


# Create a simulated genomic matrix.
chromosomes =  ['1'] * 50 + ['2'] * 50
starts = np.hstack([range(0, 500, 10), 
                    range(0, 500, 10)])
ends = starts + 10

# Create a dataframe.
data = pd.DataFrame({
    'chromosome': chromosomes,
    'start': starts,
    'end': ends,
    'sample_1': np.hstack([np.random.randn(50), 
                           np.random.randn(50) + 10]),
    'sample_2': np.hstack([np.random.randn(20) + 15, 
                           np.random.randn(30) + 0,
                           np.random.randn(50) + -10]),
    'sample_3': np.hstack([np.random.randn(25) + 10, 
                           np.random.randn(25) + 0,
                           np.random.randn(50) + -5])
})
```

```python
from genopandas import GenomicMatrix

matrix = GenomicMatrix(data.set_index(['chromosome', 'start', 'end']))
matrix.head()
```

```python
matrix.plot_heatmap()
```

```python
resampled = matrix.resample(bin_size=20, start=0)
resampled.plot_heatmap()
```

```python
matrix.plot_sample('sample_1', markersize=5)
```
