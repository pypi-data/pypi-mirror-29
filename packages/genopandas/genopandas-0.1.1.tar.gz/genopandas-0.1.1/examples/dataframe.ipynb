# Construction

Creating an instance from an existing dataframe:

```python
from genopandas import GenomicDataFrame

df = pd.DataFrame.from_records(
    [('1', 10, 20, 'a'),
     ('2', 10, 20, 'b'),
     ('2', 30, 40, 'c')],
    columns=['chromosome', 'start', 'end', 'name'])

df = df.set_index(['chromosome', 'start', 'end'])

gdf = GenomicDataFrame(df)

gdf.head()
```

Reading from delimited files using from_csv:

```python
gdf = gpd.GenomicDataFrame.from_csv(
    '../data/example.gtf', sep='\t', 
    names=('contig', 'source', 'feature', 'start', 'end', 'score',
           'strand', 'frame', 'attributes'), 
    index_col=['contig', 'start', 'end'])

gdf.head()
```

Reading from specialized file formats:

```python
import genopandas as gpd

gdf = gpd.GenomicDataFrame.from_gtf('../data/example.gtf')
gdf.head()
```
